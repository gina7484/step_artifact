```python
import numpy as np


batch = 16
# idx_list = [(1456,2805,2820),(1455,821,836),(1454,985,1000)]
# idx_list = [(987,3349,3364),(987,3275,3290),(986,3181,3196)]
idx_list = [(480,3799,3814),(479,2477,2492),(479,1063,1078)]

dir = f"/home/ginasohn/step_tl/dynamic_par/azure_trace/b{batch}/"

mean_list = []
std_list = []
for stdev,start, end in idx_list:
    file_path = f"{dir}/conv_stdev{stdev:04d}_{start:04d}_{end:04d}.npy"
    raw_np_arr = np.load(file_path)
    np_arr_int = raw_np_arr.astype(np.int64)
    num_token_list = np_arr_int.tolist()
    print(f"num_token_list: {num_token_list}")

    mean = np.mean(num_token_list)
    std = np.std(num_token_list)
    mean_list.append(mean)
    std_list.append(std)

print(f"geomean of mean: {np.prod(mean_list)**(1/len(mean_list))}")
print(f"geomean of std: {np.prod(std_list)**(1/len(std_list))}")
```
